#**CHI-SQUARE TEST FOR INDEPENDENCE - ASSOCIATION BETWEEN DEVICE TYPE AND CUSTOMER SATISFACTION**

**INTRODUCTION**

Mizzare Corporation is a company that sells smart home devices. Recently, they collected customer satisfaction data for two types of devices:Smart Thermostats and Smart Lights.

The goal is to check whether the type of device purchased is related to the customer satisfaction level. In other words, we want to know if there is a statistical association between these two variables.

To test this, we will use the **Chi-Square Test for Independence**, a statistical method used to determine if two categorical variables are related.

**PROBLEM STATEMENT :**

We will perform the Chi-Square test to determine if there is a significant association between:
1. Device Type (Smart Thermostat, Smart Light)
2. Customer Satisfaction Level (Very Satisfied, Satisfied, Neutral, Unsatisfied, Very Unsatisfied)

**STEP-1 : STATING THE HYPOTHESIS FOR THE MODEL**

**NULL HYPOTHESIS(H0) :** There is no association between the device type and customer satisfaction level.

**ALTERNATIVE HYPOTHESIS(H1) :** There is an association between device type and customer satisfaction level.


**STEP-2 : PYTHON CODING**

```python
#Importing the Libraries we need
import numpy as np #for creating arrays
import pandas as pd #for creating dataframes
from scipy.stats import chi2_contingency #runs the chi2 test for testing the independence
from scipy.stats import chi2 #calculating the critical value for the distribution
from scipy import stats
#Creating an array and dataframe of the given data to know the actual data from testing
#data = np.array([
    #[50, 70],   # Very Satisfied
    #[80, 100],  # Satisfied
    #[60, 90],   # Neutral
    #[#30, 50],   # Unsatisfied
    #[20, 50]    # Very Unsatisfied
#])
#df = pd.DataFrame(
    #data,
    #columns=["Smart Thermostat", "Smart Light"],
    #index=["Very Satisfied", "Satisfied", "Neutral", "Unsatisfied", "Very Unsatisfied"]
#)
#print("Observed Frequencies (Contingency Table):\n")
#print(df, "\n")
```

```python
data = [
    ["Satisfaction", "Smart Thermostat", "Smart Light", "Total"],
    ["Very Satisfied", 50, 70, 120],
    ["Satisfied", 80, 100, 180],
    ["Neutral", 60, 90, 150],
    ["Unsatisfied", 30, 50, 80],
    ["Very Unsatisfied", 20, 50, 70],
    ["Total", 240, 360, 600]
]

# Convert the list 'data' into a pandas DataFrame
df = pd.DataFrame(data[1:], columns=data[0])  # Use data[1:] for data rows and data[0] for column names

df
```

```python
#Stating the Hypothesis
print("Hypothesis:\n")
print("H₀: There is no association between device type and customer satisfaction level.")
print("H₁: There is an association between device type and customer satisfaction level.\n")
```

```python
df.iloc[0:5,1:3]
```

```python
chi2_stat, p_value, dof, expected = chi2_contingency(df.iloc[0:5,1:3])

print(chi2_stat,',',p_value,',',dof)
```

```python
alpha=0.05
critical_value = stats.chi2.ppf(1 - alpha, df=dof)

print(critical_value)
```

```python
if p_value < alpha:
    decision = "Reject H0"
else:
    decision = "Accept H0"
decision
```

```python
if p_value <= alpha:
    print("\nConclusion: There IS a statistically significant association between device type and customer satisfaction level.")
else:
    print("\nConclusion: There is NO statistically significant association between device type and customer satisfaction level.")
```

#**CONCLUSION**

Based on the results of the Chi-Square test for independence, the observed Chi-Square statistic is less than the critical value, and the p-value is greater than the significance level (α = 0.05). Therefore, we fail to reject the null hypothesis (H₀). This indicates that there is no statistically significant association between the type of smart home device (Smart Thermostat vs. Smart Light) and the level of customer satisfaction. In other words, the variation in customer satisfaction levels is likely due to chance rather than the type of device purchased.
