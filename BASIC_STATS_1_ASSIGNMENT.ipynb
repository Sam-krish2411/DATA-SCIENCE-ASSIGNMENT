# **DESCRIPTIVE ANALYTICS FOR NUMERICAL COLUMN**

```python
import pandas as pd
```

```python
df = pd.read_excel('sales_data.xlsx')
numerical_cols = df.select_dtypes(include='number').columns
print(f"Numerical Columns Found: {list(numerical_cols)}")

# Calculate descriptive statistics
mean_vals = df[numerical_cols].mean()
median_vals = df[numerical_cols].median()
mode_vals = df[numerical_cols].mode().iloc[0]
std_dev_vals = df[numerical_cols].std()

summary_stats = pd.DataFrame({
    'Mean': mean_vals,
    'Median': median_vals,
    'Mode': mode_vals,
    'Standard Deviation': std_dev_vals
})


print("\nDescriptive Statistics Summary:\n")
print(summary_stats)
```

# **DATA VISUALISATION**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_excel('sales_data.xlsx')

print(df.head())
```

```python
# HISTOGRAM
numerical_cols = df.select_dtypes(include='number').columns

for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color='Red')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
#BOXPLOT
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='red')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
#
categorical_cols = df.select_dtypes(include='object').columns

```

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar', color='darkgreen')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
df = pd.read_excel('sales_data.xlsx')

numerical_cols = df.select_dtypes(include='number').columns
df_numerical = df[numerical_cols]


for col in numerical_cols:
    sns.histplot(df[col], kde=True)
    plt.title(f'Original Distribution: {col}')
    plt.show()

scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df_numerical), columns=numerical_cols)

for col in numerical_cols:
    sns.histplot(df_standardized[col], kde=True, color='lightblue')
    plt.title(f'Standardized Distribution: {col}')
```
