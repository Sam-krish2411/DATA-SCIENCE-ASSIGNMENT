**STEP-1 : DESCRIPTIVE ANALYTICS FOR NUMERICAL COLUMN


```python
#computing the basic statistical measures for the numerical column
import pandas as pd

df = pd.read_csv("sales_data_with_discounts.csv")
numerical_cols = df.select_dtypes(include='number').columns
print(f"Numerical Columns Found: {list(numerical_cols)}")

# Calculate descriptive statistics
mean_vals = df[numerical_cols].mean()
median_vals = df[numerical_cols].median()
mode_vals = df[numerical_cols].mode().iloc[0]
std_dev_vals = df[numerical_cols].std()

summary_stats = pd.DataFrame({
    'Mean': mean_vals,
    'Median': median_vals,
    'Mode': mode_vals,
    'Standard Deviation': std_dev_vals
})
```

STEP-02 - DATA VISUALISATION FOR NUMERICAL COLUMN

```python
#HISTOGRAM
import matplotlib.pyplot as plt
import seaborn as sns

for col in numerical_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True, bins=30,palette='Set1')
    plt.title(f"Distribution of {col}")
    plt.show()
```

**INTERPRETATION**

1. Volume is moderately right-skewed.

2. Average Price and Sales Values are heavily right-skewed, confirming presence of a few very high transactions.

3. Discount Rate is closer to normal distribution, centered around ~15%.

```python
#BOXPLOT
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col],palette='Set2')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

**INTERPRETATIONS**

1. The boxplots show that most variables, such as Discount Rate (%), have a relatively **small interquartile range (IQR)**, indicating consistency in discounts. In contrast, variables like Average Price and Total Sales Value have **a very wide IQR**, reflecting high variability across transactions.

2. **Several outliers** are present in Discount Amount and Net Sales Value, suggesting unusually large discounts or sales in certain cases.

3. The distributions of Average Price and Total Sales Value appear **right-skewed**, showing that while most values are low to moderate, a few extreme high values pull the data upward.

Overall, the plots highlight variability, skewness, and the presence of extreme values in sales data.

```python
#BAR CHARTS FOR CATEGORICAL DATA
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col],palette='Set3')
    plt.title(f"Category Distribution: {col}")
    plt.xticks(rotation=45)
    plt.show()
```
